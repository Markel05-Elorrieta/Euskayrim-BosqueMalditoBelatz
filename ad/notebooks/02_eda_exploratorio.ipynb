Importamos las herramientas de análisis (Pandas, gráficos) y cargamos el archivo CSV en la memoria para empezar a trabajar:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv('bosque_maldito_limpio.csv')

print(df.shape)
display(df.head())
```

Se cargan pandas para el manejo de tablas, matplotlib y seaborn para la creación de gráficos, y numpy para operaciones matemáticas. Estas son las herramientas fundamentales. Además, observamos las primeras filas del Dataset

Revisamos la estructura técnica de los datos:

```python
df.info()

print(df.isnull().sum().sum())

display(df.describe())
```

En esta celda verificamos los tipos de datos (principalmente float64 e int64) y confirmamos la ausencia de valores nulos, lo cual es un requisito crítico de la rúbrica para el bloque de limpieza de datos. Asegurar que no hay datos faltantes garantiza que el entrenamiento posterior de la Red Neuronal sea estable y no produzca errores inesperados

Aplicamos ahora la lógica del juego para encontrar datos imposibles (como vida 100) en los héroes activos:

```python
heroes = ["olenthero", "thorgin", "amalyria", "basajorn", "lamyreth", "sugarth"]
errores_detectados = []

if not df['dificultad'].between(1, 10).all():
    errores_detectados.append("Hay dificultades fuera de rango")

for h in heroes:
    col_equipo = f"{h}_en_equipo"
    col_vida = f"{h}_vida"

    if col_vida in df.columns:
        stats_activos = df[df[col_equipo] == 1][col_vida]

        if not stats_activos.between(1, 10).all():
            fuera_rango = stats_activos[~stats_activos.between(1, 10)]
            errores_detectados.append(f"{h} tiene vida fuera de rango")

if len(errores_detectados) == 0:
    print("Todo correcto")
else:
    print("SE DETECTARON ANOMALÍAS:")
    for e in errores_detectados:
        print(e)
```

Se realiza un análisis de distribución para asegurar que los atributos de los héroes han sido normalizados y que no existen sesgos que puedan confundir el entrenamiento de la Red Neuronal

Ahora, contamos cuántas veces se ganó vs. perdió:

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='exito', data=df, palette='viridis')
plt.title('Distribución de éxito')
plt.xlabel('Resultado')
plt.ylabel('Cantidad')

total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()
```

 Esto nos sirve para saber si el juego es demasiado difícil y si la IA tendrá suficientes ejemplos de victoria para aprender

Creamos ahora un mapa de calor para ver qué estadísticas (fuerza, agilidad, etc.) están más relacionadas matemáticamente con ganar la partida:

```python
cols_numericas = df.select_dtypes(include=[np.number]).columns
df_corr = df[cols_numericas]

plt.figure(figsize=(20, 15))
sns.heatmap(df_corr.corr(), annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de calor de correlaciones')
plt.show()

print(df_corr.corr()['exito'].sort_values(ascending=False).head(10))
```

El mapa de calor visualiza las correlaciones entre todas las variables numéricas, permitiendo identificar rápidamente relaciones. La lista impresa debajo muestra las 10 columnas con mayor correlación positiva con el 'éxito', lo que sugiere qué atributos pueden influir más en el resultado de la partida.

Y, por último, mostramos qué héroes se usan más:

```python
popularidad = df[[f'{h}_en_equipo' for h in heroes]].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=popularidad.index, y=popularidad.values, palette='magma')
plt.title('Frecuencia de Héroes')
plt.xticks(rotation=45)
plt.show()
```

Aquí, calculamos y visualizamos la popularidad de cada héroe. El gráfico de barras muestra cuántas veces cada héroe fue incluido en un equipo. Esto es útil para identificar patrones de uso, como qué personajes son los más o menos elegidos, y puede indicar posibles sesgos en los datos que el modelo de IA podría aprender
